##이미지 보기


```python
import matplotlib.pyplot as plt
from torchvision import datasets

test_DS = datasets.MNIST(root="data/", train=False, download=True)
#손글씨 데이터
```

```python
print(test_DS.data.shape)
print(test_DS.data[0])
```

```python
plt.imshow(test_DS.data[0]) #image show
```

```python
plt.imshow(test_DS.data[1], cmap = "gray")
plt.colorbar()
```

```python
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(test_DS.data[i], cmap="gray")
```

```python
plt.figure(figsize=(7,4))
for i in range(6):
    plt.subplot(2,3,i+1, xticks=[], yticks=[])
    plt.imshow(test_DS.data[i],cmap="gray")
```

```python
plt.figure(figsize=(7,4))
for i in range(6):
    plt.subplot(2,3,i+1, xticks=[], yticks=[])
    plt.imshow(test_DS.data[i], cmap="gray")
    plt.title(test_DS.targets[i].item(), color="g")
```

```python
import torch

im = test_DS.data[0]
plt.figure(figsize=(20,15))
plt.imshow(im, cmap='gray')
plt.colorbar()
for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        plt.text(j,i,im[i,j].item(),
                 horizontalalignment="center",
                 color="k" if im[i,j] > torch.max(im)/1.5 else "w")
```
